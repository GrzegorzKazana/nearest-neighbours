```python
import tensorflow as tf
import numpy as np
import pandas as pd
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pathlib

import transform
```

```python
BATCH_SIZE = 32
```

```python
train_dir = pathlib.Path.cwd() / "train"
# test_dir = pathlib.Path.cwd() / "test"
validation_dir = pathlib.Path.cwd() / "validation"

df = pd.read_csv(pathlib.Path.cwd() / "data" / "labels.csv")
df
```

```python
training_df = df.loc[df.task2_class != "validation"]
training_df
```

```python
validation_df = df.loc[df.task2_class == "validation"]
validation_df
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    channel_shift_range=20,
    horizontal_flip=True,
    fill_mode="reflect"
)
validation_datagen = ImageDataGenerator(
    rescale=1./255
)
```

```python
cols = [name for name in training_df.columns if name[0].isupper()]
train_generator = train_datagen.flow_from_dataframe(training_df, 
                                                    directory=str(train_dir), 
                                                    x_col="filename", 
                                                    y_col=cols,
                                                    target_size=(422,422),
                                                    class_mode="raw", 
                                                    batch_size=32, 
                                                    shuffle=True)
print(cols)
validation_generator = validation_datagen.flow_from_dataframe(validation_df,
                                                              directory=str(validation_dir),
                                                              x_col="filename", 
                                                              y_col=cols,
                                                              class_mode="raw",
                                                              batch_size=32)
```

### Model building

```python
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.backend import clear_session
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Dense, Dropout, Flatten, AveragePooling2D
from tensorflow.keras import initializers, regularizers
```

```python
clear_session()

height, width = 422, 422

conv_base = InceptionV3(
    weights='imagenet', 
    include_top=False, 
    input_shape=(height, width, 3)
)
conv_base.trainable = False
for layer in conv_base.layers:
    layer.trainable = False

print('Summary')
print(conv_base.summary())
```

```python
x = conv_base.output
x = AveragePooling2D(pool_size=(8, 8))(x)
x - Dropout(0.4)(x)
x = Flatten()(x)
x = Dense(128, 
          activation='relu', 
          kernel_initializer=initializers.he_normal(seed=None), 
          kernel_regularizer=regularizers.l2(.0005))(x)
x = Dropout(0.5)(x)

x = Dense(64,
          activation='relu', 
          kernel_initializer=initializers.he_normal(seed=None))(x)
x = Dropout(0.25)(x)
predictions = Dense(len(cols),  
                    kernel_initializer="glorot_uniform", 
                    activation='softmax')(x)

print('Stacking New Layers')
model = Model(inputs = conv_base.input, outputs=predictions)
```

```python
print('Compile model')
opt = SGD(momentum=.9)
model.compile(
    loss='binary_crossentropy',
    optimizer=opt,
    metrics=['accuracy']
)
```

```python
print('Start training!')
history = model.fit_generator(
    train_generator,
    epochs=100,
    steps_per_epoch=20,
    shuffle=True,
    validation_data=validation_generator,
    validation_steps=50
)

print('Saving Model')
model.save("nsfw." + str(width) + "x" + str(height) + ".h5")
```
